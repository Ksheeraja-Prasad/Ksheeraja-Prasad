```python
#Change in production of rice,wheat and vegetable over time

import matplotlib.pyplot as plt
%matplotlib inline

data.plot(x='year',y=['rice_production','wheat_production','vegetable_production'])
plt.title('Global Food Production')
plt.ylabel('Production in million metric tons')
plt.show()
```

```python
#Countries that are highest and lowest in rice production
highest_rice=data.sort_values('rice_production',ascending=False).head(5)
lowest_rice=data.sort_values('rice_production',ascending=True).head(5)

fig,ax1 = plt.subplots(1,1, figsize=(10, 5))

ax1.bar(highest_rice['Country'], highest_rice['rice_production'], color='green')
ax1.bar(lowest_rice['Country'], lowest_rice['rice_production'], color='red')
ax1.set_title('Rice Production')
```

```python
#Countries that are highest and lowest in wheat production
highest_wheat=data.sort_values('wheat_production',ascending=False).head(5)
lowest_wheat=data.sort_values('wheat_production',ascending=True).head(5)

fig,ax1 = plt.subplots(1,1, figsize=(10, 5))

ax1.bar(highest_wheat['Country'], highest_wheat['wheat_production'], color='green')
ax1.bar(lowest_wheat['Country'], lowest_wheat['wheat_production'], color='red')
ax1.set_title('Wheat Production')
```

```python
#Countries that are highest and lowest in vegetable production
highest_vegetable=data.sort_values('vegetable_production',ascending=False).head(5)
lowest_vegetable=data.sort_values('vegetable_production',ascending=True).head(5)

fig,ax1 = plt.subplots(1,1, figsize=(10, 5))

ax1.bar(highest_vegetable['Country'], highest_vegetable['vegetable_production'], color='green')
ax1.bar(lowest_vegetable['Country'], lowest_vegetable['vegetable_production'], color='red')
ax1.set_title('Vegetable Production')
```
